```r
require(readr)
require(ggplot2)
require(GGally)
        
```

```r
# This is classic data set from a paper by Ronald Fisher in 1936. It contains morphologic features of 
# Iris flowers of three related species
# load into a dataframe   
df <- readr::read_csv('iris.data',col_names=c('SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'))
# just checking if the data was loaded correctly
head(df)
```

```r
# We have four attributes for each flower and 150 data points in total.
summary(df)
```

```r
# An interesting observation is that two of the flower species are not separable from
# each other
ggplot(df,aes(x=SepalLengthCm,y=SepalWidthCm,color=Species)) + geom_point()
```

```r
# The pair plot illustrates again that the flower species are not linearly separable.
# It show the bivariate relation between each pair of measurements.
ggpairs(df, ggplot2::aes(colour=Species, alpha=0.4))
#plotmatrix(df[,1:4],aes(color=Species))
```

```r
ggplot(df, aes(x=Species, y=PetalLengthCm, fill=Species)) + geom_boxplot()
```

```r
ggplot(df, aes(x=Species, y=PetalWidthCm, fill=Species)) + geom_boxplot()
```
